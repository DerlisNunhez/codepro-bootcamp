```python
#Crear un diccionario
mi_contacto = {'nombre': 'Fulana',
               'email':'prueba@gmail.com',
               'domicilio':'asuncion'}

print(mi_contacto)

#Acceder a una clave del diccionario
mi_contacto['email']
print(mi_contacto)

#Modificar un elemento del diccionario
mi_contacto['email']='daniel@gmail.com'
print(mi_contacto)

#Usar metodos del diccionario
print(mi_contacto.keys()) #devuelve todas las claves del diccionario
print(mi_contacto.values()) #devuelve todos las valores del diccionario
print(mi_contacto.items()) #devuelve el par
print(mi_contacto.get('nombre')) #devuelve el valor

#Anhadir mas datos al diccionario
mi_contacto['ciudad'] = 'Asuncion'
print(mi_contacto)

#Eliminar el ultimo elemento(clave-valor) del diccionario
mi_contacto.popitem()
print(mi_contacto)

#Eliminar datos del diccionario
del mi_contacto['email']
print(mi_contacto)

#los diccionarios son iterables
for valores in mi_contacto.values():
    print(f'Valores{valores}')
print('')

for claves in mi_contacto.keys():
    print(f'Claves{claves}')
print('')

for clave,valor in mi_contacto.items():
    print(f'Clave y Valores{clave, valor}')
print('')
```

```python
nombre = 'Daniel'
apellido = 'Nunhez'
ciudad = 'Asuncion'
cel = '0985'
empresa = 'Penguin'
tarjeta = {'nombre':nombre,
           'apellido':apellido,
           'ciudad':ciudad,
           'cel':cel,
           'empresa':empresa}

print('TARJETA')
for clave, valor in tarjeta.items():
    print(f'{clave}:{valor}')

while True:
    print('\nElija una opcion(escribir solo numero)\n1. Crear datos nuevos\n2. Modificar datos\n3. Eliminar datos\n4. Salir')
    datos = int(input())

    if datos == 1:
        clave = input('Que dato desea agregar?')
        valor = input('Que deberia decir?')
        tarjeta[clave]=valor
        print('TARJETA')
        for clave, valor in tarjeta.items():
            print(f'{clave}:{valor}')
    elif datos == 2:
        clave = input('Que dato desea modificar?')
        valor = input('Que deberia decir?')
        tarjeta[clave]=valor
        print('TARJETA')
        for clave, valor in tarjeta.items():
            print(f'{clave}:{valor}')
    elif datos == 3:
        clave = input('Que dato desea eliminar?')
        del tarjeta[clave]
        print('Eliminado')
        print('TARJETA')
        for clave, valor in tarjeta.items():
            print(f'{clave}:{valor}')
    elif datos == 4:
        print('Gracias por consultar')
        break
    else:
        print('Elija del 1 al 4\n')
```

```python
class Superheroe:
    #atributo de clase
    especie = 'superhumano'

    #Definir metodo constructor para definir sus atributos de instancia
    def __init__(self,nombre):
        self.nombre=nombre

    #metodo
    def rayos_laser(self):
        print('Tiro rayos laser')

    def volar(self,km):
        print(f'Estoy volando a {km}')

    def alzar_objetos(self,objeto,kg):
        print(f'Puedo alzar un {objeto} de {kg}kg')


#instanciar objetos
super1 = Superheroe('Superman')
super1.volar(120)

super2 = Superheroe('Flash')
super2.alzar_objetos('edificio',2000)
print(f'Me llamo {super2.nombre}')

```

```python
class Coche:
    tipo = 'vehiculo'

    def __init__(self, modelo, posicion = 0):
        self.modelo = modelo
        self.posicion = posicion

    def acelerar(self, metros):
        self.posicion += metros
        print(f'El vehiculo modelo {self.modelo} acelera y avanza {metros}m. Nueva posicion {self.posicion}m')

    def girar(self,direccion):
        print(f'El {self.modelo} avanza y gira a la direccion {direccion} ')


auto1 = Coche('supra',14)
auto1.acelerar(47)
auto1.girar('izquierda')
```

```python
class Mascota:
    especie = 'ornitorrinco'

    def __init__(self, nombre):
        self.nombre = nombre

    def hablar(self,palabras):
        self.palabra = palabras
        print(f'Cuando habla dice {palabras}')

    def movimiento(self,moverse):
        self.mover = moverse
        print(f'Cuando se mueve, lo hace {moverse}')

animal = Mascota('Perry')
print(f'Se llama {animal.nombre} el {animal.especie}')
animal.hablar('Prrrr')
animal.movimiento('sigilosamente para que nadie lo vea')
```

```python
class AnimalDeGranja:
    #atributo de clase
    tipo = 'animal de granja'

    #atributo de instancia
    def __init__(self,nombre, color, sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido
    
    #metodos (acciones de la clase)
    def hablar(self):
        print(f'Hablo y digo {self.sonido}')
    def presentarse(self):
        print(f'Hola, soy {self.nombre}')



class Gallina(AnimalDeGranja):
    huevos_por_dia = 1

    def poner_huevo(self):
        print(f'Estoy poniendo {self.huevos_por_dia}')


class Vaca(Gallina, Mascota):
    patas = 4

    def mujir(self):
        print(f'Muuuu')

gallina1 = Gallina('pepa','amarillo','co co')
gallina1.hablar()
gallina1.presentarse()

vaca1 = Vaca('Lola','negro','hola')
vaca1.hablar()
vaca1.poner_huevo()
vaca1.hablar()
```

```python
class Madre:
    
    def __init__(self, nombre, edad, color_pelo):
        self.nombre = nombre
        self.edad = edad
        self.color_pelo = color_pelo

class Hija1(Madre):

    def presentacion(self):
        print(f'Yo me llamo {self.nombre} mi edad es {self.edad} y mi pelo es de color {self.color_pelo}')
        
class Hija2(Hija1):

    def presentacion(self):
        print(f'Mi nombre es {self.nombre} tengo {self.edad} y mi cabello es de color {self.color_pelo}')

hermana_mayor = Hija1('Luz',20,'naranja')
hermana_menor = Hija2('Sol',14,'rojo')

hermana_mayor.presentacion()
hermana_menor.presentacion()

```

```python
class Heroe:
    persona1 = 'Con superpoderes buenos'
    def __init__(self,nombre,poder,motivacion):
        self.nombre = nombre
        self.poder = poder
        self.motivacion = motivacion

    def demo_fuerza(self):
        print(f'Mi nombre es {self.nombre}, yo puedo {self.poder} y peleo por {self.motivacion}')


class Villano(Heroe):
    persona2 = 'Con superpoderes malos'

heroe1= Heroe('Metroman','volar, tengo superfuerza, rayos laser','cuidar la ciudad del mal')
villano1 = Villano('Megamente','ser muy fastidioso pero super inteligente','que no quiero que me molesten los heroes')

heroe1.demo_fuerza()
villano1.demo_fuerza()


```
