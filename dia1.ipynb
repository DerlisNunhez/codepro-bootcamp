```python
def compra(elementox,altox, largox, anchox):
    print(f'El/La {elementox} que queres comprar tiene:\n{altox} de alto\n{largox} de largo\n{anchox} de ancho')

elemento = input('Que producto queres comprar?')
alto = input('Cuanto de alto deberia tener?')
largo = input('Cuanto de largo deberia tener?')
ancho = input('Cuanto de ancho deberia tener?')

compra(elemento, alto, largo, ancho)
```
