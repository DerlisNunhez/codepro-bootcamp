```python
def saludar():
    print('Bienvenido')
    print('Hora de programar')

contra_co = 'perro'
contra_ingre = 0

contra_ingre = input('Por favor ingrese su contraseña')
while contra_ingre != contra_co:
    contra_ingre = input('Contraseña incorrecta, intente de vuelta')
    print('Pista: Es un animal')
else:
    saludar()
```

```python

def letras(letra_a,letra_b):
    if letra_a == letra_b:
        print('Las letras son iguales')
    else:
        print('Las letras son diferentes')

letra1 = input('Por favor coloque la primera letra')
letra2 = input('Por favor coloque la segunda letra')

letras(letra1,letra2)
```

```python
def sumar(n1,n2):
    resultado_suma = n1 * n2
    return resultado_suma

resultado = sumar(3,2)
print(resultado)


def retirar_dinero(dinero):
    return f'Retirando {dinero}'

dinero_retirado = retirar_dinero(5000)
print(f'Este es el dato de {dinero_retirado}')

sumar(dinero_retirado,1000)
```

```python
palabra = input('Escriba una palabra')
numero = int(input('Escriba un numero'))

def datos(pal,num):
    if len(pal) == num:
        return True
    else:
        return False
    
datos(palabra, numero)

    
```

```python
palabra = input('Por favor escriba una palabra')
letra = input('Escriba una sola letra')

def revisar(palabra, letra):
        if letra in palabra:
            return f'[{letra}]'
        else:
            return letra

revisar(palabra, letra)
```
